```python
from pytrends.request import TrendReq
pytrends = TrendReq()
```

```python
kw_list = ["Fin del Mundo", "Calentamiento Global", "Terremoto", "Tsunami"]
pytrends.build_payload(kw_list, cat=0, geo='', gprop='')
stop_queries = pytrends.interest_over_time()[kw_list]
stop_queries.head()
```

```python
stop_queries.tail()
```

*Qué significan los valores?*

De acuerdo con Google Trends, los valores son calculados en una escala de 0 a 100, donde 100 indica la popularidad máxima del término mientras que 50 o 0 significan que los términos son la mitad o carecen de nula popularidad respecto del 100. También puede deberse a qué no habían demasiados datos.

En síntesis, los números reflejan el interés de búsqueda en relación con el valor máximo de un gráfico en una región y un periodo determinados

```python
stop_queries.plot(kind='line',figsize=(12,6), xlabel='Fecha',ylabel='Interés de audiencia',title='Interés en el tiempo')
```
